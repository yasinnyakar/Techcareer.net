#### Veri Ã¼zerinde ilk dÃ¼zeltmeleri ve analizleri yaptÄ±ktan sonra boÅŸ verileri doldurma aÅŸamasÄ±na geÃ§iyoruz. Ä°lk olarak boÅŸ veri olan kolonlardaki daÄŸÄ±lÄ±mlara bakÄ±yoruz.

```python
na_cols = df.isna().sum()[
                            df.isna().sum() > 0
                        ].index 

for i in df[na_cols]:
    print(i)
    display(df[i].value_counts().sort_index())
```

#### BoÅŸ veri iÃ§eren kolonlarÄ±, kolonun ortalama deÄŸeriyle dolduruyoruz ve sonrasÄ±nda baÅŸarÄ±yla doldurulduÄŸunu kontrol ediyoruz.

```python
for i in df[na_cols]:
    df[i] = df[i].fillna(df[i].mean())
    
df.isna().sum()
```

#### Outlier analizi iÃ§in ortalama +/- 3 standart sapma yaklaÅŸÄ±mÄ±nÄ± kullanarak kontrolleri saÄŸlÄ±yoruz ve bu aralÄ±ÄŸÄ±n dÄ±ÅŸÄ±nda kalan deÄŸerleri aralÄ±ÄŸÄ±n sÄ±nÄ±rÄ±na getiriyoruz.

```python
for i in df.columns:
    upp = df[i].mean() + 3 * df[i].std()
    low = df[i].mean() - 3 * df[i].std()
    df[i] = df[i].apply(lambda x: upp if x > upp else low if x < low else x)
df.describe().T
```
