#### Veri üzerinde ilk düzeltmeleri ve analizleri yaptıktan sonra boş verileri doldurma aşamasına geçiyoruz. İlk olarak boş veri olan kolonlardaki dağılımlara bakıyoruz.

```python
na_cols = df.isna().sum()[
                            df.isna().sum() > 0
                        ].index 

for i in df[na_cols]:
    print(i)
    display(df[i].value_counts().sort_index())
```

#### Boş veri içeren kolonları, kolonun ortalama değeriyle dolduruyoruz ve sonrasında başarıyla doldurulduğunu kontrol ediyoruz.

```python
for i in df[na_cols]:
    df[i] = df[i].fillna(df[i].mean())
    
df.isna().sum()
```

#### Outlier analizi için ortalama +/- 3 standart sapma yaklaşımını kullanarak kontrolleri sağlıyoruz ve bu aralığın dışında kalan değerleri aralığın sınırına getiriyoruz.

```python
for i in df.columns:
    upp = df[i].mean() + 3 * df[i].std()
    low = df[i].mean() - 3 * df[i].std()
    df[i] = df[i].apply(lambda x: upp if x > upp else low if x < low else x)
df.describe().T
```

#### Tüm özellliklerin ortalaması olacak şekilde yeni bir değişken oluşturuyoruz.

#### Kategorik değişkenimiz olan position_id'yi get_dummies methoduyla(kategorik değişkenimizi dummy değişkene döndürüyoruz) one-hot encoding(çünkü yukarıda 1 ve 0 olarak dönüşüm yaptık) edip modelimize uygun hale getiriyoruz.

#### Modelimizin daha sağlıklı çalışması için tüm özellik değişkenlerimizi scale ediyoruz.

```python
from sklearn.preprocessing import StandardScaler

df["avg"] = df.iloc[:, 1:-1].mean(axis=1)

scaler = StandardScaler()

for i in df.columns:
    if i not in ["position_id", "target"]:
        df[i] = scaler.fit_transform(df[i].values.reshape(-1,1))
        
df = pd.get_dummies(df, columns = ["position_id"])
df
```
